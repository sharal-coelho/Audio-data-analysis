To get Statistical Features from wav file use below code

```python
import numpy as np
import scipy
```

```python
freqns = np.fft.fftfreq(signal.size)
print(freqns)
```

```python
def freq_details(freqns):
    mean = np.mean(freqns)
    std = np.std(freqns) 
    minv = np.amin(freqns) 
    maxv = np.amax(freqns) 
    median = np.median(freqns)
    mode = scipy.stats.mode(freqns)[0][0]
    skew = scipy.stats.skew(freqns)
    kurt = scipy.stats.kurtosis(freqns)
    q1 = np.quantile(freqns, 0.25)
    q3 = np.quantile(freqns, 0.75)
    iqr = scipy.stats.iqr(freqns)
    
    return [mean, std, minv, maxv, median, mode, skew, kurt, q1, q3,  iqr]

#get frequency details
mean, std, minv, maxv, median, mode, skew, kurt, q1, q3,  iqr=freq_details(freqns)
```

```python
print("mean :", mean)
print("std :", std)
print("minv :", minv)
print("maxv :", maxv)
print("median :", median)
print("mode :", mode)
print("skew :", skew)
print("kurt :", kurt)
print("q1 :", q1)
print("q3 :", q3)
print("iqr :", iqr)
```
